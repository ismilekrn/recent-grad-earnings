```python
recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

## Scatter Plots

```python
%matplotlib inline
ax1 = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
%matplotlib inline
ax2 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

```python
%matplotlib inline
ax3 = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
```

```python
%matplotlib inline
ax4 = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

```python
%matplotlib inline
ax5 = recent_grads.plot(x='Women', y='Median', kind='scatter')
```

```python
columns = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize=(16,5))

for r in range(0,4):
    ax = fig.add_subplot(1,4,r+1) #r=1,2,3,4
    ax = recent_grads[columns[r]].plot(kind='hist', rot=30)
```

```python
columns = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize=(5,10))

for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3) #r=1,2,3,4
    ax = recent_grads[columns[r]].plot(kind='hist', rot=30)
```
